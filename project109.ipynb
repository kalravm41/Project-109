```
import statistics
import random
import plotly.figure_factory as px
import plotly.graph_objects as go
import pandas as pd
import csv

df = pd.read_csv('data.csv')
dice_result = df['attempt'].to_list()
# dice_result = []

for i in range(0,100):
  dice1 = random.randint(1,6)
  dice2 = random.randint(1,6)
  
  dice_result.append(dice1 + dice2)

mean = sum(dice_result)/len(dice_result)
print('mean is {}'.format(mean))

median = statistics.median(dice_result)
print('median is {}'.format(median))

mode = statistics.mode(dice_result)
print('mode is {}'.format(mode))

sd = statistics.stdev(dice_result)
print('standard deviation is {}'.format(sd))

firstsdst, firstsdend = mean-sd, mean+sd
secondsdst, secondsdend = mean-(2*sd), mean+(2*sd)
thirdsdst, thirdsdend = mean-(3*sd), mean+(3*sd)

listOfdata1 = [result for result in dice_result if result > firstsdst and result < firstsdend]
listOfdata2 = [result for result in dice_result if result > secondsdst and result < secondsdend]
listOfdata3 = [result for result in dice_result if result > thirdsdst and result < thirdsdend]

print('{}% of data lies within 1 standard deviation'.format(len(listOfdata1)*100/len(dice_result)))
print('{}% of data lies within 2 standard deviation'.format(len(listOfdata2)*100/len(dice_result)))
print('{}% of data lies within 3 standard deviation'.format(len(listOfdata3)*100/len(dice_result)))

fig = px.create_distplot([dice_result],['results'], show_hist = False)
fig.add_trace(go.Scatter(x= [mean,mean], y= [0,0.2], mode= 'lines', name= 'mean'))

fig.add_trace(go.Scatter(x= [firstsdst, firstsdst], y= [0,0.2], mode= 'lines', name= 'Standard Deviation 1 start'))
fig.add_trace(go.Scatter(x= [firstsdend, firstsdend], y= [0,0.2], mode= 'lines', name= 'Standard Deviation 1 end'))

fig.add_trace(go.Scatter(x= [secondsdst, secondsdst], y= [0,0.2], mode= 'lines', name= 'Standard Deviation 2 start'))
fig.add_trace(go.Scatter(x= [secondsdend, secondsdend], y= [0,0.2], mode= 'lines', name= 'Standard Deviation 2 end'))

fig.add_trace(go.Scatter(x= [thirdsdst, thirdsdst], y= [0,0.2], mode= 'lines', name= 'Standard Deviation 3 start'))
fig.add_trace(go.Scatter(x= [thirdsdend, thirdsdend], y= [0,0.2], mode= 'lines', name= 'Standard Deviation 3 end'))

fig.show()
```
